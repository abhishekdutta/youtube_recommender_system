```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import tokenize
import csv
import itertools
import re
```

```python
link='commentsFile.csv'

input1 = open(link, 'r')

lines = itertools.islice(input1,1,None)
reader = csv.reader(lines)

full_len = len(list(reader))

print(full_len)

input1.close()
```

```python
words = set(nltk.corpus.words.words())
```

```python
texts=[]

link="commentsFile.csv"

with open(link, newline='', encoding='utf8') as f:
    lines = itertools.islice(f,1,None)
    reader = csv.reader(lines)
    
    for row in reader:
        
        check1=" ".join(w for w in nltk.wordpunct_tokenize(row[3]) \
                if w.lower() in words or not w.isalpha())
        
        comments_arr=check1.split(";")
        
        strip1=map(lambda x: re.sub(r"http\S+",'',x), comments_arr)
        strip2=list(strip1)
        
        strip3=map(lambda x: re.sub(r"a href",'',x), strip2)
        strip4=list(strip3)
        
        strip5=map(lambda x: x.strip(), strip4)
        strip6=list(strip5)
        strip7=filter(None,strip6)
        texts.append(list(strip7))
        
'''
texts = ["This is good, catchy, and hillarious.", 
"This is intellectual, awesome, and hillarious!", 
"This is very intellectual, awesome, and hillarious.",  
"This is VERY intellectual, awesome, and HILLARIOUS.",  
"This is VERY intellectual, awesome, and HILLARIOUS!!!",
"This is VERY intellectual, really awesome, and INCREDIBLY Hillarious!!!",
"The movie was fine.",        
"The movie was a one time watch.",  
"The plot was amazing, but the characters are boring and the dialog is not great.", 
"A really bad, horrible book.",       
"At least it isn't a horrible book.", 
":) and :D",     
"",             
"Today was cool",     
"This was dope!",    
"This was pathetic!",    
"This kinda was creepy! But there were a lot of lol moments" 
]
'''

print(list(texts))

#Examples include:-

# positive sentence example
# punctuation shows emphasis 
# sample positive words 
# emphasis for CAPS 
# combination of words
# positive words and punctuation (higher score)
# positive sentence
# qualified as positive sentence
# mixed negation sentence
# negative sentence with negative words
# negated negative sentence 
# emoticons 
# empty string 
# positive slang   
# negative slang with punctuation 
# negative slang with punctuation 
# mixed sentiment example 
```

```python
#title = "Trailer of Knock once"

# Adding title to the feature space after comments 

with open(link, newline='', encoding='utf8') as f:
    lines = itertools.islice(f,1,None)
    reader = csv.reader(lines)
    
    i=0
    for row in reader:
        texts[i].append(row[1])
        i+=1
        
print(texts)
```

```python
#description = "This is a trailer of the new upcoming horror movie!!!!"

# Adding description to the feature space after comments and title 

with open(link, newline='', encoding='utf8') as f:
    lines = itertools.islice(f,1,None)
    reader = csv.reader(lines)
    
    i=0
    for row in reader:
        texts[i].append(row[2])
        i+=1
        
print(texts)
```

```python
# Calculting sentiment scores

analyser = SentimentIntensityAnalyzer()

for i in range(0,len(texts)):
    for review in texts[i]:
        print(review)
        sentiment_score = analyser.polarity_scores(review)
        for i in sorted(sentiment_score):
            print('{0}: {1}, '.format(i, sentiment_score[i]), end='')
        print()
    print()
```

```python
# Compound score is normalized score for each sentence
# Average compound score is calculated for the entire video
# Range is from -1 to +1

avg=[]
for i in range(0,len(texts)):
    sum1=0
    for review in texts[i]:
        print(review)
        sentiment_score = analyser.polarity_scores(review)
        sum1=sum1+sentiment_score['compound']
        print('compound: ' + str(sentiment_score['compound']))

        print()
    
    avg_score=sum1/len(texts[i])
    avg.append(avg_score)
    print(avg_score)
    
    print()
```

```python
# We classify videos as negative (<0), neutral (=0) and positive (>0)

print(len(avg))
category_list=[]

count_pos=0
count_neg=0
count_neutral=0

for i in range(0,len(avg)):
    if avg[i]<0:
        print('negative')
        count_neg+=1
        category_list.append('negative')
    elif avg[i]==0:
        print('neutral')
        count_neutral+=1
        category_list.append('neutral')
    else:
        print('positive')
        count_pos+=1
        category_list.append('positive')
print(category_list)
```

```python
read_link='commentsFile.csv'

# Header
list1=['Id', 'Title', 'Description', 'Comments (; delimited list)', 'Sentiment Percentage', 'Sentiment Category']
tot_list=[]
tot_list.append(list1)

with open(read_link, newline='', encoding='utf8') as f:
    lines = itertools.islice(f,1,None)
    reader = csv.reader(lines)
    
    row_num=0
    for row in reader:
        list2=[]
        for j in range(0,4):
            list2.append(row[j])
        list2.append(avg[row_num]*100)
        list2.append(category_list[row_num])
        tot_list.append(list2)
        row_num+=1
    print(tot_list)
```

```python
# Writing to file the predicted labels

write_link='sentiment_analysis_of_top50_videos.csv'

with open(write_link, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(tot_list)
```

```python
# Percentage of each sentiment category (Positive, Negative, Neutral)

print("Percentage of Positive videos: " + str(float(count_pos*100/full_len))+"%")
print("Percentage of Negative videos: " + str(float(count_neg*100/full_len))+"%")
print("Percentage of Neutral videos: " + str(float(count_neutral*100/full_len))+"%")
```
